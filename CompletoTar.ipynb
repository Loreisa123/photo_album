```python
#¿Es un playlist bailable?
#¿Dónde?
##¿Cómo?

-poner un parametro de restricción
    -Si es menos de 0.5, no es bailable

#Respuestas
¿Cuantás canciones son bailables usando una restricción D?
```

```python
#Generacion de pandita(Tomo solo una columna)
danza=spotify['danceability']

#Pregunta 1

#Parametro de restriccion
D=0.7

#Total de canciones
C=danza.count()

#Balderas choice
#C1=len(danza)

#Número de canciones bailables sum-sum
C_B=sum(danza>D)

#Porporción de canciones bailables en la lista
P_B=C_B/C
print('Existen '+str(C_B)+' que son bailables de un total de '+str(C))
print("El"+str(round(P_B,4)*100)+'%de las canciones son bailables según un parametro D='+str(D))
```

```python
round(P_B,4)
```

```python
#Gráfica de bailable, categorizar, frecuencia es el histograma
#Metodo Matplotlib.pyplot mejor conocido plt
plt.hist(danza,30, facecolor='pink')
#Etiquetas
plt.xlabel('Bailabilidad[ua]')
#Frecuencia de numero de canciones
plt.ylabel('Numero de canciones')
#Titulo
plt.title('Número de canciones,respecto a la bailabilidad ')
#El show se pone al final ya que es el tope de lo que se esta haciendo
plt.show()
```

```python
danza.plot(kind='hist',bins=30)
plt.xlabel('Bailabilidad[ua]')
#Frecuencia de numero de canciones
plt.ylabel('Numero de canciones')
#Titulo
plt.title('Número de canciones,respecto a la bailabilidad ')
#El show se pone al final ya que es el tope de lo que se esta haciendo
plt.show()
```

```python
#¿Como es la tendencia de la duración?
duracion=spotify['duration_ms']
m=duracion/60000
#print(str(round(m,4)))

plt.hist(m,40, facecolor='pink')
#Etiquetas
plt.xlabel('duracion[ua]')
#Frecuencia de numero de canciones
plt.ylabel('Tendencia de canciones')
#Titulo
plt.title('Duracion de canciones ')
#El show se pone al final ya que es el tope de lo que se esta haciendo
plt.show()
```

```python
#No tiene instrumentalización
#Generacion de pandita(Tomo solo una columna)
inst=spotify['instrumentalness']

#Pregunta 1

#Parametro de restriccion
D=0.3

#Total de canciones
C=inst.count()

#Número de canciones bailables sum-sum
C_B=sum(inst<D)

#Porporción de canciones bailables en la lista
P_B=C_B/C

print('Existen '+str(C_B)+' canciones sin instrumentalizacion '+str(C))
print("El"+str(round(P_B,4)*100)+'%')

plt.hist(inst,40, facecolor='pink')
#Etiquetas
plt.xlabel('inst[ua]')
#Frecuencia de numero de canciones
plt.ylabel('Canciones')
#Titulo
plt.title('Tendencia de canciones sin instrumentalización ')
#El show se pone al final ya que es el tope de lo que se esta haciendo
plt.show()
```
