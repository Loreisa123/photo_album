```python
#Muestra los países con esperanza de vida femenina mayor al promedio general y sus datos
OCDE2=OCDE2.set_index("Indicator")
```

```python
EV=OCDE2.loc["Life expectancy"]
EV[["Unit","Value","Inequality"]]
```

```python
EVP=EVPP.loc["Women"]
EVP[["Value","LOCATION"]]
```
